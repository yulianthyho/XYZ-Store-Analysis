#Import Library

```python
# import library
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from datetime import datetime
from datetime import timedelta
```

#Load the Dataset

```python
url_1 = 'https://docs.google.com/spreadsheets/d/1Q3F8E1kFn4FkhREijcVCuxXLblAHsO_LHOqAokg9qO4/edit#gid=75363367'
url_sheet_trf_1 = url_1.replace('/edit#gid=', '/export?format=xlsx&gid=')

# load the order_dataset
df_order = pd.read_excel(url_sheet_trf_1)
pd.set_option('display.max.columns',None)
# get the overview of the dataset
df_order.head()
```

```python
url_2 = 'https://docs.google.com/spreadsheets/d/1Q3F8E1kFn4FkhREijcVCuxXLblAHsO_LHOqAokg9qO4/edit#gid=2082975771'
url_sheet_trf_2= url_2.replace('/edit#gid=', '/export?format=xlsx&gid=')


# load the order_payments_dataset
df_returned = pd.read_excel(url_sheet_trf_2)
pd.set_option('display.max.columns',None)

#get the overview of the dataset
df_returned.head()
```

# Merge the Dataset

```python
#data merge
df = df_order.merge(df_returned, how='left', on='order_id')
df
```

# Cleaning the Data

## Handling Missing Values

```python
#checking missing values of dataset
df.isna().sum()
```

We will drop 63 rows in column Product Base Margin (COGS) and fillna Status with Not returned

```python
#fillna status with no returned
df['Status']= df['Status'].fillna('Not Returned')
```

```python
#because we can't subset the null data , we drop the null data
df.dropna(inplace=True)
```

```python
df.isna().sum()
```

Our dataset is now clean from missing values

## Handling Duplicate Data

```python
df.duplicated().sum()
```

Our dataset is clean from duplicated data

## Removing Irrelevant Values

```python
#removing irrelevant column
df= df.drop(columns=['Row ID'])
```

```python
df.info()
```

```python
#change order id from float to object
df['order_id'] = df['order_id'].astype(str)
df['order_id'] = df['order_id'].str.split('.', expand = True)[0]
```

```python
#change order quantity from float to int
df['order_quantity'] = df['order_quantity'].astype(int)
```

# Descriptive Statistics

Adding Calculated Field to see Profit Margin , The formula is (Profit / Discounted Sales) * 100. We use discounted sales here, in order to better describe the company's profit because it takes into account the discount

```python
#adding calculated field
df['gross_profit_margin'] = (df.Profit / df['discounted_sales'])
df.head(5)
```

To get more full describe about our profit and sales performance, we consider to not remove outlier from our dataset and consider where status is Not Returned

```python
#get the dataset where status is not returned
df1 = df.loc[(df['Status']=='Not Returned')]
```

```python
# Get descriptive statistics summary
df1.describe(include = "all")
```

**Key Findings:**

- On average our Gross Profit Margin is 44%, which is considerable quite high
- Top customer is Corporate with best selling product is Paper from Office Supplies category
- Top shipmode used is Regular Air

# Exploratory Data Analysis

**1. Which Category is Best Selling and Most Profitable?**

```python
# Group sales, profit and quantity by category
category_analysis = pd.DataFrame(df1.groupby(['product_category'])[['discounted_sales', 'Profit', 'order_quantity']].sum())
pd.set_option('display.float_format',  '{:,}'.format) # set the display of numbers so that they are not abbreviated
category_analysis = round(category_analysis,2)                          # set the display of numbers to be rounded to 2 digits after the comma
category_analysis
```

- Technology is Best Selling and it's good to know that this category is the Most Profitable too. 
- Although Furniture makes the seconds position from Technology, it is the least profitable and quantity sold are at a minimum too.
- Office Supplies sells the most in terms of quantity as it is relatively cheap product.


**2. What are the Best Selling and Most Profitable Sub-Category?**

```python
# Group by sub-category
subcat_analysis = pd.DataFrame(df1.groupby(['product_sub_category'])[['discounted_sales', 'Profit', 'order_quantity']].sum())
pd.set_option('display.float_format',  '{:,}'.format) # set the display of numbers so that they are not abbreviated
subcat_analysis = round(subcat_analysis,2)                          # set the display of numbers to be rounded to 2 digits after the comma
subcat_analysis
```

```python
# Sort by descending order according to sales
subcat_sales = pd.DataFrame(subcat_analysis.sort_values('Profit', ascending = False))
subcat_sales
```

- The most profitabe products are Office Machines and Telephone and Communication
- The least profitable is Labels, Scissors , Rules and Trimmers, and Rubber Bands. All from Offices Supplies category

**3. Which Customer Segment is Most Profitable?**

```python
segment_analysis = pd.DataFrame(df1.groupby(['customer_segment'])[['Profit']].sum())
pd.set_option('display.float_format',  '{:,}'.format) # set the display of numbers so that they are not abbreviated
segment_analysis = round(segment_analysis,2)                          # set the display of numbers to be rounded to 2 digits after the comma
segment_analysis
```

Corporate segment is most profitable, followed by Home Office and Consumer Segment


**4. Which Region is the Most Profitable?**

```python
region_analysis = pd.DataFrame(df1.groupby(['Region'])['Profit'].sum().reset_index())
pd.set_option('display.float_format',  '{:,}'.format) # set the display of numbers so that they are not abbreviated
region_analysis = round(region_analysis,2)                          # set the display of numbers to be rounded to 2 digits after the comma
region_analysis

```

```python
# Sort by descending order according to profi
region_analysis = pd.DataFrame(region_analysis.sort_values('Profit', ascending = False))
region_analysis
```

Top region that contribute the most profit is West, while Nunavut become the region with the least number of profit.
